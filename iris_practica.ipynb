```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
```

```python
iris = sns.load_dataset('iris')
# pd.set_option('display.max_rows',None)
iris.head()
```

```python
total = iris.mean(axis=0)
total = pd.DataFrame(total)
total.columns=['mean']
plt.bar(total.index, total['mean'])
plt.show()
```

```python
iris[iris['species']=='setosa'].describe()
```

```python
iris[iris['species']=='versicolor'].describe()
```

```python
iris[iris['species']=='virginica'].describe()
```

```python
data= iris.groupby(['species']).mean()
data = data.transpose()
data
```

```python
n = len(data.index)
x = np.arange(n)
width = 0.25

plt.bar(x - width, data.setosa, width=width, label='setosa')
plt.bar(x, data.versicolor, width=width, label='versicolor')
plt.bar(x + width, data.virginica, width=width, label='virginica')
plt.xticks(x, data.index)
plt.legend(loc='best')
plt.show()
```
